```python
import torch.nn as nn
from torch.optim import lr_scheduler
import torch
import numpy as np
```

```python

from cifar_dl_dt_helper import get_dataloaders_sizes_classes
from CifarModel1 import Cifar_model1
from pytorch_training2 import Pytorch_training_helper
```

```python
from pytorch_model_helper import get_n_params
```

```python
data_dir = './data/cifar10'
result_folder = "reek_cifar_3Models_grad_and_weights/"
```

```python
batch_size = 64
num_epochs = 100
```

```python
from CifarModel1 import Cifar_model1
from CifarModel2 import Cifar_model2
from CifarModel7 import Cifar_model7
```

```python
cifar_model1 = Cifar_model1()
cifar_model2 = Cifar_model2()
cifar_model7 = Cifar_model7()
```

```python
print("Model Parameters")
print("Model 1:- ",get_n_params(cifar_model1))
print("Model 2:- ",get_n_params(cifar_model2))
print("Model 7:- ",get_n_params(cifar_model7))
```

```python
dataloaders,dataset_sizes,class_names = get_dataloaders_sizes_classes(data_dir,batch_size)
py_train_helper = Pytorch_training_helper(dataloaders,dataset_sizes,batch_size,result_folder,
                                          flag_grad=1,
                                          flag_weights=1,
                                          flag_frobenius_norm=1)
```

```python
step = 0.0004   
gamma_lr_scheduler = 0.1
criterion = nn.CrossEntropyLoss()
weight_decay = 1e-4
```

```python
print("Model 1")
# cifar_model1 = Cifar_model1()

# optimizer = optim.SGD(model.parameters(), lr=0.001)
optimizer1 = torch.optim.Adam(cifar_model1.parameters(),lr = step,weight_decay=weight_decay)

exp_lr_scheduler1 = lr_scheduler.StepLR(
    optimizer1, step_size=10, gamma=gamma_lr_scheduler
)
model_name1 = "CifarModel1"

cifar_model1,grad_norm_per_epoch1,train_losses1,val_losses1,train_acc1,val_acc1,model_wts_epoch1,sensitivity1 = py_train_helper.train_model(cifar_model1,criterion,optimizer1,exp_lr_scheduler1,model_name1,num_epochs=num_epochs)
```

```python
print("Model 2")
# cifar_model1 = Cifar_model1()

# optimizer = optim.SGD(model.parameters(), lr=0.001)
optimizer2 = torch.optim.Adam(cifar_model2.parameters(),lr = step,weight_decay=weight_decay)

exp_lr_scheduler2 = lr_scheduler.StepLR(
    optimizer2, step_size=10, gamma=gamma_lr_scheduler
)
model_name2 = "CifarModel2"

# cifar_model1,grad_norm_per_epoch1,train_losses1,val_losses1,train_acc1,val_acc1,model_wts_epoch1,sensitivity1 = py_train_helper.train_model(cifar_model1,criterion,optimizer,exp_lr_scheduler,model_name1,num_epochs=num_epochs)

cifar_model2,grad_norm_per_epoch2,train_losses2,val_losses2,train_acc2,val_acc2,model_wts_epoch2,sensitivity2  = py_train_helper.train_model(cifar_model2,criterion,optimizer2,exp_lr_scheduler2,model_name2,num_epochs=num_epochs)
```

```python
print("Model 7")

optimizer7 = torch.optim.Adam(cifar_model7.parameters(),lr = step,weight_decay=weight_decay)

exp_lr_scheduler7 = lr_scheduler.StepLR(
    optimizer7, step_size=10, gamma=gamma_lr_scheduler
)
model_name7 = "CifarModel7"
cifar_model7,grad_norm_per_epoch7,train_losses7,val_losses7,train_acc7,val_acc7,model_wts_epoch7,sensitivity7 = py_train_helper.train_model(cifar_model7,criterion,optimizer7,exp_lr_scheduler7,model_name7,num_epochs=num_epochs)
```
